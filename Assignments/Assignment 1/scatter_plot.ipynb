```python
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

eps = 1
minPts = 4
data_filename = "hepta.dat"

command = "make; ./asm " + str(eps) + " " + str(minPts) + " " + data_filename

os.system(command)

# Reading dataset
df = pd.read_csv('a.csv')

# Creating figure
fig, axs = plt.subplots(1, 2, figsize=(12, 8), subplot_kw={'projection': '3d'})

# Creating plot
plot(axs[0], df, "label")
plot(axs[1], df, "pred")

# show plot
plt.show()
```

```python
def plot(ax, df, label):
#     colors = ['blue', 'green', 'red', 'cyan', 'magenta', 'yellow', 'black', 'white']
    legend = []
    i = 1
    for pl in df[label].unique():
        fdl = df[df[label] == pl]
        ax.scatter3D(fdl.x, fdl.y, fdl.z, color="C" + str(i))
        legend.append(str(pl))
        i = i + 1
    ax.legend(legend)
    ax.set_title(label.capitalize() + " scatter plot")
```
